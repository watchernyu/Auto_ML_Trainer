```python
globPaths = ['data/c0/*.jpg','data/c1/*.jpg','data/c2/*.jpg','data/c3/*.jpg','data/c4/*.jpg','data/c2/*.jpg',
            'data/c6/*.jpg','data/c7/*.jpg','data/c8/*.jpg','data/c9/*.jpg']
savePaths = ['data/c0small/','data/c1small/','data/c2small/','data/c3small/','data/c4small/','data/c5small/',
            'data/c6small/','data/c7small/','data/c8small/','data/c9small/']
writeLimit = 100 #read every 100 images, then write to disk

for folderIndex in range(10):
    filelist = glob(globPaths[folderIndex])
    processedImages = []
    i = 0
    for filename in filelist:
        newImg = misc.imread(filename)
        newImg = misc.imresize(newImg,(64,64))
        misc.imsave(savePaths[folderIndex]+str(i)+'.jpg', newImg)
        i+=1
        if i%100==0:
            print(folderIndex,i)

#by now we finished creating small images
```
